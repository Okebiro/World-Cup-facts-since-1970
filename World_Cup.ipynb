```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
World_Cup = pd.read_csv('Worldcupmen.csv')
World_Cup.info()
```

```python
World_Cup1 = World_Cup.dropna(axis=1)

World_Cup1.head()
```

```python
World_Cup1.columns
```

```python
World_Cup1.head()
```

```python
World_Cup2 = World_Cup1.drop(['tournament_name','group_name','replayed', 'replay', 'team_code','opponent_code','team_prior_matches', 'team_prior_win_rate',
       'team_prior_goals_scored_avg', 'team_prior_goals_conceded_avg',
       'opp_prior_matches', 'opp_prior_win_rate', 'opp_prior_goals_scored_avg',
       'opp_prior_goals_conceded_avg', 'h2h_prior_matches',
       'h2h_prior_win_rate','team_curr_form_pts_avg',
       'team_curr_goals_scored_avg', 'team_curr_goals_conceded_avg',
       'opp_curr_form_pts_avg', 'opp_curr_goals_scored_avg',
       'opp_curr_goals_conceded_avg','match_time','home_team', 'away_team','penalties_for',
       'penalties_against','yellow_cards_h1', 'yellow_cards_h2',
       'red_cards_h1', 'red_cards_h2','is_host'], axis=1)
World_Cup2["Total_Goals"] = World_Cup2["goals_for"] + World_Cup2["goals_against"]
World_Cup2.head()
```

Dropping duplicate games

```python
World_Cup2["pair_key"] = World_Cup2[["team_name", "opponent_name"]].apply(
    lambda row: tuple(sorted(row)),
    axis=1
)

World_Cup2 = (
    World_Cup2.drop_duplicates(subset=["match_date", "pair_key"], keep="first")
      .drop(columns="pair_key"))
```

Number of Worldcups played

```python
Winners = World_Cup2[World_Cup2['stage_name'] == 'final']
len(Winners)
```

How many countries have won world cup since 1970

```python
Winners['team_name'][Winners.outcome == 'Win'].value_counts().plot(kind='bar',color='#5EF056')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('World Cup Winners Since 1970')
plt.xticks(rotation=45, ha='right')
```

Top 10 matches the most goals and its stage

```python

MostGoals = World_Cup2.sort_values(by="Total_Goals", ascending=False).head(10)
MostGoals
```

Top five finals with most goals

```python
FinalsGoals = Winners.sort_values(by="Total_Goals", ascending=False).head(5)
FinalsGoals
```
